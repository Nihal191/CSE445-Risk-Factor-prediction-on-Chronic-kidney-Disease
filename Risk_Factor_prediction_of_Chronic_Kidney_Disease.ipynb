Handling Null Values

Here we got only one row containing null value and one row contains discrete value so we drop those two rows for handling the null values.

```python
data1 = dataset.drop(dataset.index[0])
data2 = data1.drop(dataset.index[1])
```

```python
data2.isnull().sum() 
```

```python
#converting the categorical values in to Numerical using LabelEncoder.
for column in data2.columns:
  if data2[column].dtype == np.number:
    continue
  data2[column] = LabelEncoder().fit_transform(data2[column])
```

```python
data2
```

# **Correlation Matrics**

```python
#Correlation plot
sns.set(rc = {'figure.figsize':(25, 15)})
sns.heatmap(data2.corr(), annot=True, cmap="YlGnBu")
```

```python
#columns we need to retain or keep
columns_to_train = ['sg','al','rbc','class','su','pc','pcc','ba','bgr','bu','hemo','pcv','rbcc','wbcc','htn','grf']
#Drop the other columns of the dataset except the columns_to_retain
data = data2.drop([col for col in data2.columns if not col in columns_to_train],axis = 1)
```

```python
data # after converting Categorical data into numerical data.
```

```python
data_x = pd.get_dummies(data)      # convert categorical variables into dummy variables
data_x
```

```python
tar_Y = data['class']  #Separating our target column and storing it into a variable.
tar_Y
```

```python
X =data.drop(["class"], axis = 'columns')  # Dropping the traget column from the dataset and storing the other columns in a variable.
X.head(10)
```

# **Splitting the Dataset into train(70%) and test (30%)**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, tar_Y, test_size = 0.3, random_state = 20) #using train_test_split function to divide the dataset into 70% for train and 30% for test.
```

```python
print(X_train.shape,Y_train.shape,X.shape,tar_Y.shape,X_test.shape,Y_test.shape)
```

```python
total_column = ['sg','al','rbc','su','pc','pcc','ba','bgr','bu','hemo','pcv','rbcc','wbcc','htn','grf']
len(total_column)
```

# **Feature Scalling**

```python
scale = MinMaxScaler() #using MinMaxScaler for converting the whole dataset columns into a fixed range of 0 t0 1.
```

```python
scale.fit(X_train[total_column])
```

```python
X_train[total_column] = scale.transform(X_train[total_column])
X_test[total_column] = scale.transform(X_test[total_column])
```

```python
X_train.head(10)
```

```python
Y_train.head(10)
```

Here according to our target variable class column the output have to ckd or notckd. We can see that its more looks like binary classification problem. For this reasons we choose this algorithms like Support Vector Machine, Naive Bayes Classifier, Logistic Regression.

# **SVM(Support Vector Machine) Algorithm**

```python
clf = svm.SVC() 
```

```python
# defining parameter range
param_grid = {'C': [0.1, 1], 
              'gamma': [1, 0.1],
              'kernel': ['linear']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, Y_train)
```

```python
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test)
  
# print classification report
print(classification_report(grid_predictions,Y_test))
print("Train Accuracy: ", accuracy_score(Y_train, grid.predict(X_train)))
print("Test Accuracy: ", accuracy_score(Y_test, grid_predictions))
```

# **Confusion Matrix**

```python
sns.set(rc = {'figure.figsize':(5, 5)})
cm1 = confusion_matrix(Y_test, grid_predictions)
sns.heatmap(cm1, cmap = "Reds", annot = True)
```

# **Naive Bayes Classifier Algorithm**

```python
classifier_predictions = classifier.predict(X_test)
  
# print classification report
print(classification_report(Y_test, classifier_predictions))
print("Train Accuracy: ", accuracy_score(Y_train, classifier.predict(X_train)))
print("Test Accuracy: ", accuracy_score(Y_test, classifier_predictions))
```

# **Logistic Regression Algorithm**

```python
l1_ratio = [float(x) for x in np.linspace(start = 0, stop = 1, num = 10)]
penalty = ['l1', 'l2', 'elasticnet']
solver = ['liblinear', 'lbfgs', 'newton-cg', 'sag', 'saga']
```

```python
random_grid = {'l1_ratio': l1_ratio,
               'penalty': penalty,
               'solver': solver}
```

```python
print(random_grid)
```

```python
lr_clf = LogisticRegression()
lr_clf=RandomizedSearchCV(estimator=lr_clf,param_distributions=random_grid,n_iter=100,cv=3,verbose=2,random_state=100,n_jobs=-1)
```

```python
lr_clf.fit(X_train,Y_train)
```

```python
lr_clf.best_params_
```
